## Google GenAI Client

```python
import os
```

```python
api_key = os.getenv("Google_API_KEY")
if not api_key:
    api_key = input("Enter your google api key")
    os.environ["GOOGLE_API_KEY"]=api_key
    if not api_key:
        raise EnvironmentError("Google API key not found")
```

```python
from google import genai

gemini_client = genai.Client(api_key=os.getenv("GOOGLE_API_KEY"))

response = gemini_client.models.generate_content(
    model="gemini-2.0-flash",
    contents=["you are a helpful assistant","what is the capital of the moon?"]
)

response.text
```

## Groq Client

```python
api_key = os.getenv("GROQ_API_KEY")
if not api_key:
    api_key = input("enter your groq api key")
    os.environ["GROQ_API_KEY"]=api_key
    if not api_key:
        raise EnvironmentError("Groq API key not found")
```

```python
from groq import Groq

client = Groq(
    api_key=os.environ.get("GROQ_API_KEY"),
)

response = client.chat.completions.create(
    messages=[{"role":"user","content":"what is the capital of the moon?"}],
    model="llama-3.3-70b-versatile"   
)

print(response.choices[0].message.content)
## OpenAI Client

```
